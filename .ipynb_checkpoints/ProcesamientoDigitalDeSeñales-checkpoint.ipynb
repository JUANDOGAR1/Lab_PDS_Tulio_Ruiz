## Introducción 
El procesamiento de **Señales sísmicas** desempeña un papel fundamental en el análisis y comprensión de los fenómenos geofísicos, especialmente en el estudio de terremotos y eventos sísmicos. Los terremotos generan señales sísmicas que contienen información valiosa sobre la propagación de ondas y las características de los eventos sísmicos. Sin embargo, estas señales a menudo están contaminadas por ruido y otras interferencias, lo que dificulta su análisis y la extracción de información relevante.

En este laboratorio, se aborda el uso de filtros FIR (Respuesta finita al impulso) como una herramienta fundamental en el procesamiento de señales sísmicas. Los filtros FIR son una clase de filtros digitales cuya respuesta al impulso es de duración finita, y se caracterizan por ser lineales e invariantes en el tiempo. Estos filtros se utilizan para mejorar la calidad de las señales sísmicas, eliminando o reduciendo el ruido no deseado y resaltando las características de interés.

### Diseño de Filtros con Respuesta Finita al Impulso (FIR)

## 1 Diseño de Filtros con Respuesta Finita al Impulso (FIR)
### 1.1.Procedimiento
#### 1.1.1 Implemente un filtro FIR con las siguientes características:
    Rechaza–banda, fc1=1kHz, fc2=4kHz, fs=16kHz, BW=200Hz, AdB=0

```python
#Consideraciones iniciales 
fc1=1000.0
fc2=4000.0
fs=16000.0 # frec de muestreo
Adb=0 # Ganancia del filtro en decibeles
BW=200 # Ancho de banda en la banda de tansicion

#Normalizan las frecuencias de corte, y se pasan a rad/s
wc1 = 2*np.pi*fc1/fs # frecuencia de corte 1 normalizada en radianes
wc2 = 2*np.pi*fc2/fs # frecuencia de corte 2 normalizada en radianes

#Normalizar ancho de banda
bwn=2*np.pi*BW/fs

# orden estimado del filtro
M=int(4/bwn) 
print('Orden del filtro: ', M)

#Vector del filtro
n = np.arange(-M,M)

# Respuesta al impulso del filtro ideal rechazabanda sin enventanado
h4 = +(wc1/np.pi)* np.sinc(wc1*(n)/np.pi)-(wc2/np.pi)* np.sinc(wc2*(n)/np.pi) 
h4[n==0]=1-((wc2-wc1)/np.pi)  #Rechazabanda truncado
```

```python
N = 512 # Numero de puntos de la FFT
win= signal.hamming(len(n)) # funcion ventana para eliminar el fenomeno de Gibbs
h4_H=h4*win # Multiplico la respuesta ideal por la ventana
A=np.sqrt(10**(0.1*Adb))
h4_H=h4_H*A # Ganancia del filtro
w4_H,Hh4_H= signal.freqz(h4_H,1,whole=True, worN=N) # Respuesta en frecuencia del filtro enventanado
# Gráfica respuesta en frecuencia con ventana Hamming 
plt.figure(figsize=(12,5))
plt.stem(n+M,h4_H)
plt.title('Filtro con Hamming',fontsize=18)
plt.xlabel("n",fontsize=14)
plt.ylabel('$H_4$',fontsize=14)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot((w4_H-np.pi)*fs/(2*np.pi),np.abs(np.fft.fftshift(Hh4_H)))
plt.title('Respuesta en frecuencia con ventana Hamming',fontsize=18)
plt.xlabel("$f (Hz)$",fontsize=14)
plt.ylabel("$|H1(\omega)| $",fontsize=14)
plt.grid()
plt.show()
```

## Diseño de Filtros FIR por Método de Muestreo en Frecuencia
El diseño de filtros FIR (Finite Impulse Response) es una tarea fundamental en el procesamiento de señales, ya que permite modificar o eliminar componentes no deseadas de una señal. Entre los métodos más utilizados para el diseño de filtros FIR se encuentra el método de muestreo en frecuencia.

El método de muestreo en frecuencia es una técnica ampliamente empleada para obtener filtros FIR con características deseables en el dominio de la frecuencia. Permite especificar de manera precisa la respuesta en frecuencia deseada, incluyendo la atenuación y ganancia en diferentes bandas, así como la forma de la curva de respuesta.
1. Tomamos $N$ muestras en intervalos de $\frac{KF_{s}}{N}$ con $K=0,1,...,\frac{N-1}{2}$
2. Los coeficientes del filtro $h(n)$ pueden obtenerse con la IDFT, así:
$$h(n)=\frac{1}{N}\sum_{k=1}^{\frac{N}{2}-1}|H(k)|\cos(\frac{2\pi k(n-\alpha)}{N})+H(0)$$
Donde $\alpha=\frac{N-1}{2}$
* Para $N$ impar el límite superior de la suma será $\frac{N-1}{2}$
3. * **Muestreo tipo 1** <br>
 Se toma un número adecuado de muestras de tal forma que se cumpla: 
 $$f_{k}=k\frac{F_{s}}{N}; k=0,1,...,\frac{N-1}{2}$$
   * **Muestreo tipo 2** <br>
   Tomar las muestras de tal forma que se cumpla:
  $$f_{k}=(k+\frac{1}{2})\frac{F_{s}}{N}; k=0,1,...,\frac{N-1}{2}$$

###  Diseñe un filtro FIR pasa-bandas con respuesta en fase lineal usando el método de muestreo en frecuencia, tal que: 
- Frecuencia de corte 1: 3700 Hz.
- Frecuencia de corte 2: 4300 Hz.
- Fs=13 kHz
- N=13 

```python
# Filtro FIR PASA BANDA por muestreo en frecuencia
Fs_fm=13000 # Frecuencia de muestreo
N=13 #orden del filtro 
fc1=3700 # Frecuencia de corte 1
fc2=4300 # Frecuencia de corte 2

M = int(Fs_fm/N) # Es para los pasos de k-> k(1khz) 
print("Se tiene muestras cada: ",M,"Hz")
K =1 # limite de la sumatoria

if N%2 ==0:
    K=int(N/2 -1)# para un N par
else:
    K=int((N-1)/2) #Para un N impar


alfa = int((N-1)/2) #Constante

hk = [] #Coeficientes de mi ventana h para encontrar la ecuacion
n =np.arange(0,K+1)

# Lleno el vector de coeficientes hk para multiplicar
for i in range(K+1):
    if i*M < fc1 or i*M>fc2:
        hk.append(0)
    else:
        hk.append(1)
        
print("El vector H(k):",hk)

#se crea una funcion symbolica 
h = 1
n = sy.Symbol("n")

for i in range(1,int(K)):
    h+=2*hk[i]*sy.cos((2*np.pi*i*(n-alfa))/N)

h=(1/N)*h

#Array para sustituir valores de n de 0 a N/2 
hn_fm = np.array([])
for i in range(N):
    #Reemplazo la n completa para obtener los coeficientes del filtro
    hn_fm= np.append(hn_fm,float(h.subs(n,i)))

w_fm,Hh_fm = sp.freqz(hn_fm,1, whole = True, worN=1024) #Respuesta en frecuencia

for i in range(N//2):
    print("h({})".format(i),"= h({})".format(N-i-1)," = ",np.round(hn_fm[i],5))
print("h(6) =",np.round(hn_fm[6],5))
```

```python
#Grafica de respusta al impulso del filtro
plt.figure(figsize=(15,7))
plt.stem(hn_fm)
plt.title("Respuesta al impulso del filtro",fontsize=20)
plt.xlabel("Muestras",fontsize=15)
plt.ylabel("Amplitud",fontsize=15)
plt.grid()
plt.show()
```

```python
# Gráfica en frecuencia del filtro 
plt.figure(figsize=(15,7))
plt.title('Respuesta en frecuencia del filtro pasa-banda por Muestreo en frecuencia', fontsize=20)
plt.plot((w_fm-np.pi)*Fs_fm/(2*np.pi), np.abs(np.fft.fftshift(Hh_fm)), label='$Filtro$')
plt.vlines(fc1, 0, 1.2, color='r', lw=2, linestyle='--',label='$F_{c1}$')
plt.vlines(fc2, 0, 1.2, color='g', lw=2, linestyle='--',label='$F_{c2}$')
plt.vlines(3000, 0, 1, color='b', lw=2, linestyle='--',label='$K=3$')
plt.vlines(5000, 0, 1, color='b', lw=2, linestyle='--',label='$K=5$')
plt.xlabel("$Frecuencia [Hz]$", fontsize=15)
plt.ylabel("$|H2(\omega)|$", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

## Aplicación del filtro FIR a los datos sísmicos:

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generación de una señal sísmica de ejemplo
sampling_rate = 1000  # Tasa de muestreo en Hz
duration = 10  # Duración de la señal en segundos
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
frequencies = [10, 50, 100]  # Frecuencias de las componentes sísmicas
amplitudes = [1, 0.5, 0.2]  # Amplitudes de las componentes sísmicas
signal_sismic = np.sum([amplitudes[i] * np.sin(2 * np.pi * frequencies[i] * t) for i in range(len(frequencies))], axis=0)

# Diseño del filtro FIR
cutoff_freq = 30  # Frecuencia de corte en Hz
filter_length = 101  # Longitud del filtro

def design_fir_filter(cutoff_freq, filter_length):
    nyquist_freq = 0.5 * sampling_rate  # Frecuencia de Nyquist
    taps = signal.firwin(filter_length, cutoff_freq, fs=sampling_rate, window='hamming')
    return taps

filter_taps = design_fir_filter(cutoff_freq, filter_length)

# Aplicación del filtro FIR a la señal sísmica
filtered_signal = signal.lfilter(filter_taps, 1, signal_sismic)

# Visualización de las señales originales y filtradas
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal_sismic)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal sísmica original')

plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal sísmica filtrada')

plt.tight_layout()
plt.show()
```
